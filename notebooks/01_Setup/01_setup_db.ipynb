# Create Catalog for all the files and tables 

```python
%sql

CREATE CATALOG IF NOT EXISTS Brazilian_E_Commerce;
```

```python
%sql

USE CATALOG brazilian_e_commerce
```

# Create schemas for medalion architecture

```python
%sql

create schema if not exists brazilian_e_commerce.bronze;
create schema if not exists brazilian_e_commerce.silver;
create schema if not exists brazilian_e_commerce.gold;

-- add a row schema for the raw data
create schema if not exists brazilian_e_commerce.row_data;
CREATE VOLUME IF NOT EXISTS brazilian_e_commerce.row_data.row;
```
